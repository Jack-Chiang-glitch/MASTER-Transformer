```python
import torch
from torch.amp import autocast
from model import MASTER
import pandas as pd
```

```python
# 創建模型
model = MASTER(
    market_dim=21,
    feature_dim=185
)

# 放到 GPU
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# 假資料
batch = 16
# [batch, stocks, time, features]
x = torch.randn(batch, 1822, 8, 185, device=device)
m = torch.randn(batch, 21, device=device)

# AMP版 forward
with autocast('cuda', dtype=torch.float16):
    out = model(x, m)


print(out.shape)
```
